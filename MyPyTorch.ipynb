Some test on image

```python
from detecto.core import Model
from detecto import utils, visualize
import numpy as np 


loadedModel=Model.load('model_weights_val.pth',['boar'])

image1 = utils.read_image('/content/drive/MyDrive/detecto/MoreBoars/Boar_110.jpg')

labels, boxes, scores = loadedModel.predict_top(image1)
predictions = loadedModel.predict_top(image1)

print(labels, boxes, scores)
print(predictions)

thresh=0.95
filtered_indices=np.where(scores>thresh)
filtered_scores=scores[filtered_indices]
filtered_boxes=boxes[filtered_indices]
num_list = filtered_indices[0].tolist()
filtered_labels = [labels[i] for i in num_list]
visualize.show_labeled_image(image1, filtered_boxes, filtered_labels)

#loadedModel.get_internal_model()




# Visualize module's helper functions

#visualize.show_labeled_image(image1, boxes, labels)



```
